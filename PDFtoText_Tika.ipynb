## Directory

The code below walks through a directory and converts the PDFS to text files. Code was found online, and modifications were made in order process and clean text.

Source: https://gist.github.com/nadya-p/373e1dc335293e490d89d00c895ea7b3

```python
def extract_text_from_pdfs_recursively(dir):
    for root, dirs, files in os.walk(dir):
        for file in files:
            path_to_pdf = os.path.join(root, file)
            [stem, ext] = os.path.splitext(path_to_pdf)
            if ext == '.pdf':
                print("Processing " + path_to_pdf)
                pdf_contents = parser.from_file(path_to_pdf)
                # convert to raw
                raw_text = r'{}'.format(pdf_contents['content'])
                # replace new lines with spaces
                raw_text = raw_text.replace("\n"," ")
                # replace double new lines with spaces
                raw_text = raw_text.replace("\n\n" , " ")
                # replace tabs with spaces
                raw_text = raw_text.replace("\t"," ")
                path_to_txt = stem + '.txt'
                with open(path_to_txt, 'w') as txt_file:
                    print("Writing contents to " + path_to_txt)
                    txt_file.write(raw_text)


if __name__ == "__main__":
    extract_text_from_pdfs_recursively(os.getcwd())
```
